## LLMIP Pipeline for Grid (Stateless) Experiment

This notebook runs the full LLMIP 4-phase pipeline:

1. **Phase 1**: Train - LLM analyzes training data
2. **Phase 2**: Predict - LLM makes predictions on test data
3. **Phase 3**: Rulebook - LLM extracts decision rules
4. **Phase 4**: Replicability - Fresh LLM reproduces predictions using only the Rulebook

The key output: **Replicability Score** = can a fresh LLM reproduce predictions using only the Rulebook?

## 1. Setup

```python
import os
import json
import pandas as pd
import numpy as np
from pathlib import Path
import time
import re
from dotenv import load_dotenv
load_dotenv()

import requests

PROJECT_DIR = Path("/home/l1zle/LLMIP")
DATA_DIR = PROJECT_DIR / "data" / "prepared"
RESULTS_DIR = PROJECT_DIR / "results"
RESULTS_DIR.mkdir(exist_ok=True)

# OpenRouter API
OPENROUTER_API_KEY = os.environ.get("OPENROUTER_API_KEY")
MODEL = "z-ai/glm-4.7"
BASE_URL = "https://openrouter.ai/api/v1"

print("Setup complete!")
print(f"OPENROUTER_API_KEY: {'Set' if OPENROUTER_API_KEY else 'NOT SET'}")
print(f"Model: {MODEL}")
```

## 2. Load Data

```python
train_features = pd.read_csv(DATA_DIR / "llmip_train_features.csv", index_col=0)
test_features = pd.read_csv(DATA_DIR / "llmip_test_features.csv", index_col=0)
train_targets = pd.read_csv(DATA_DIR / "llmip_train_targets.csv", index_col=0)
test_targets = pd.read_csv(DATA_DIR / "llmip_test_targets.csv", index_col=0)

vm_cols = [c for c in train_targets.columns if c.startswith('vm_')]
train_targets = train_targets[vm_cols]
test_targets = test_targets[vm_cols]

print("=== Data Loaded ===")
print(f"Train features: {train_features.shape}")
print(f"Test features:  {test_features.shape}")
print(f"Train targets:  {train_targets.shape}")
print(f"Test targets:   {test_targets.shape}")
```

## 3. LLM Client

```python
def call_llm(prompt, system=None, max_tokens=500, temperature=0.3):
    """Call OpenRouter API."""
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {OPENROUTER_API_KEY}",
        "HTTP-Referer": "https://llmip.research",
        "X-Title": "LLMIP Research"
    }
    
    messages = []
    if system:
        messages.append({"role": "system", "content": system})
    messages.append({"role": "user", "content": prompt})
    
    payload = {
        "model": MODEL,
        "messages": messages,
        "max_tokens": max_tokens,
        "temperature": temperature
    }
    
    try:
        resp = requests.post(f"{BASE_URL}/chat/completions", headers=headers, json=payload, timeout=120)
        if resp.status_code == 200:
            result = resp.json()
            msg = result["choices"][0]["message"]
            return msg.get("content") or msg.get("reasoning") or ""
        else:
            print(f"Error: {resp.status_code} - {resp.text[:100]}")
    except Exception as e:
        print(f"Exception: {e}")
    return None

# Test
test_resp = call_llm("Say 'Test OK' in exactly those words.")
print(f"Test response: {test_resp[:50] if test_resp else 'None'}")
```

```python
def run_phase1(call_llm, train_features, train_targets, n_samples=159):
    """Phase 1: LLM analyzes training data."""
    print("\n" + "="*60)
    print("PHASE 1: TRAINING ANALYSIS")
    print("="*60)

    features_subset = train_features.iloc[:n_samples]
    targets_subset = train_targets.iloc[:n_samples]

    feature_descriptions = {
        'hour': 'Hour of day (0-23)',
        'dayofweek': 'Day of week (0=Mon, 6=Sun)',
        'month': 'Month (1-12)',
        'is_weekend': 'Weekend flag (0=weekday, 1=weekend)',
        'hour_sin': 'Cyclical hour encoding (sine)',
        'hour_cos': 'Cyclical hour encoding (cosine)',
        'month_sin': 'Cyclical month encoding (sine)',
        'month_cos': 'Cyclical month encoding (cosine)',
        'slack_vm_pu': 'Slack bus voltage magnitude (pu)',
        'slack_va_degree': 'Slack bus voltage angle (degrees)',
        'slack_bus': 'Slack bus number',
        'load_p_r1_mw': 'Active power demand in Region 1 (MW)',
        'load_q_r1_mvar': 'Reactive power demand in Region 1 (MVar)',
        'load_p_r2_mw': 'Active power demand in Region 2 (MW)',
        'load_q_r2_mvar': 'Reactive power demand in Region 2 (MVar)',
        'load_p_r3_mw': 'Active power demand in Region 3 (MW)',
        'load_q_r3_mvar': 'Reactive power demand in Region 3 (MVar)',
        'load_p_total_mw': 'Total system active power demand (MW)',
        'load_q_total_mvar': 'Total system reactive power demand (MVar)',
        'gen_solar_p_r1_mw': 'Solar generation in Region 1 (MW)',
        'gen_wind_p_r1_mw': 'Wind generation in Region 1 (MW)',
        'gen_hydro_p_r1_mw': 'Hydro generation in Region 1 (MW)',
        'gen_thermal_p_r1_mw': 'Thermal generation in Region 1 (MW)',
        'gen_solar_p_r2_mw': 'Solar generation in Region 2 (MW)',
        'gen_wind_p_r2_mw': 'Wind generation in Region 2 (MW)',
        'gen_hydro_p_r2_mw': 'Hydro generation in Region 2 (MW)',
        'gen_thermal_p_r2_mw': 'Thermal generation in Region 2 (MW)',
        'gen_solar_p_r3_mw': 'Solar generation in Region 3 (MW)',
        'gen_wind_p_r3_mw': 'Wind generation in Region 3 (MW)',
        'gen_hydro_p_r3_mw': 'Hydro generation in Region 3 (MW)',
        'gen_thermal_p_r3_mw': 'Thermal generation in Region 3 (MW)',
        'net_interchange_mw': 'Net power interchange (MW)',
    }

    feat_desc_str = "\n".join(
        f"  - {k}: {v}" for k, v in feature_descriptions.items() if k in train_features.columns
    )

    target_desc = (
        "Voltage magnitude (vm_pu) at each of 118 buses. "
        "Generator buses (typical: 1, 4, 6, 8, 10, 12, 15, 18, 24, 25, 26, 31... ) "
        "typically operate at 1.05-1.20 pu. "
        "Load buses typically operate at 0.95-1.05 pu. "
        "Normal operating range is 0.95-1.05 pu."
    )

    system_prompt = """You are an expert power systems engineer analyzing IEEE 118-bus power flow data."""

    user_prompt = f"""## Training Data Summary\n\nYou are given {n_samples} samples of IEEE 118-bus power system data.\n\n### Input Features:\n{feat_desc_str}\n\n### Target Variables (voltage magnitude in per-unit at each bus):\n{target_desc}\n\n### Your Task:\nAnalyze this data to LEARN the patterns that predict voltage magnitudes. Identify:\n1. Feature thresholds that determine voltage levels\n2. Conditional rules (IF-THEN patterns)\n3. Bus categories and their typical voltage ranges\n4. Key relationships between input features and output voltages\n\nFocus on learning DECISION RULES that can be used to predict voltages for new data.\n\n## Representative Training Samples\n\n{features_subset.to_string(max_rows=15)}\n\n## Corresponding Target Values\n\n{targets_subset.to_string(max_rows=15)}\n"""

    print(f"Sending {n_samples} samples to LLM for analysis...")
    print("(This may take a moment)")

    analysis = call_llm(system_prompt, user_prompt, temperature=0.3, max_tokens= 20192)

    if analysis:
        print(f"\nPhase 1 complete! Analysis length: {len(analysis)} chars")
        with open(RESULTS_DIR / "phase1_analysis.txt", "w") as f:
            f.write(analysis)
        print(f"Saved to: {RESULTS_DIR / 'phase1_analysis.txt'}")
    else:
        print("WARNING: Phase 1 returned empty!")

    return analysis
```

## 5. Phase 2: Predictions

```python
from pathlib import Path

def load_phase1_analysis(path=RESULTS_DIR / "phase1_analysis.txt"):
    if not path.exists():
        print(f"No Phase 1 analysis file found at {path}")
        return None
    with open(path, "r") as f:
        analysis = f.read()
    print(f"Loaded Phase 1 analysis from {path} (length={len(analysis)} chars)")
    return analysis
analysis = load_phase1_analysis()
```

```python
def parse_grid_predictions(llm_response):
    """Parse voltage predictions from LLM response."""
    voltages = {}
    if not llm_response:
        return voltages

    lines = llm_response.split('\n')
    for line_idx, line in enumerate(lines):
        match = re.search(r'Bus:?\s*(\d+)', line, re.IGNORECASE)
        if match:
            bus = match.group(1)
            voltage_match = re.search(r'(?:vm_pu|voltage[:\s]+(?:pu)?)\s*[:=]?\s*([0-9.]+)', line, re.IGNORECASE)
            if not voltage_match:
                for j in range(line_idx+1, min(line_idx+3, len(lines))):
                    voltage_match = re.search(r'([0-9]+\.[0-9]+)', lines[j])
                    if voltage_match:
                        break
            if voltage_match:
                try:
                    voltages[f'vm_{bus}'] = float(voltage_match.group(1))
                except:
                    pass
    return voltages
```

```python
# Phase 2: Bulk Parallel Predictions
print("=== Phase 2: Bulk Parallel Predictions ===")

import concurrent.futures
import threading

# Feature descriptions
feature_descriptions = {
    'hour': 'Hour of day (0-23)',
    'dayofweek': 'Day of week (0=Mon, 6=Sun)',
    'month': 'Month (1-12)',
    'is_weekend': 'Weekend flag',
    'slack_vm_pu': 'Slack bus voltage (pu)',
    'load_p_r1_mw': 'R1 load (MW)',
    'load_p_r2_mw': 'R2 load (MW)',
    'load_p_r3_mw': 'R3 load (MW)',
    'gen_solar_p_r1_mw': 'R1 solar (MW)',
    'gen_wind_p_r1_mw': 'R1 wind (MW)',
    'gen_hydro_p_r1_mw': 'R1 hydro (MW)',
    'gen_thermal_p_r1_mw': 'R1 thermal (MW)',
    'gen_solar_p_r2_mw': 'R2 solar (MW)',
    'gen_wind_p_r2_mw': 'R2 wind (MW)',
    'gen_hydro_p_r2_mw': 'R2 hydro (MW)',
    'gen_thermal_p_r2_mw': 'R2 thermal (MW)',
    'gen_solar_p_r3_mw': 'R3 solar (MW)',
    'gen_wind_p_r3_mw': 'R3 wind (MW)',
    'gen_hydro_p_r3_mw': 'R3 hydro (MW)',
    'gen_thermal_p_r3_mw': 'R3 thermal (MW)',
    'net_interchange_mw': 'Net interchange (MW)',
}

def predict_single_sample(args):
    """Predict voltages for a single test sample."""
    idx, row, analysis, lock = args
    
    feat_str = "\n".join(
        f"  {col} ({feature_descriptions.get(col, col)}): {row[col]:.4f}"
        for col in test_features.columns if col in row.index
    )
    
    system_prompt = f"""You are an expert power systems engineer. You have analyzed IEEE 118-bus power flow data and learned these patterns:

{analysis[:3000]}

Use this knowledge to predict voltage magnitudes for the new test case below."""

    user_prompt = f"""## Test Case

Timestamp: {idx}

### System Features:
{feat_str}

### Your Task:
Predict voltage magnitude (vm_pu) at each of the 118 buses. 

Format your response as:
Bus 1: X.XXX pu
Bus 2: X.XXX pu
...
Bus 118: X.XXX pu

Focus on realistic voltage values:
- Generator buses: 1.05-1.20 pu  
- Load buses: 0.95-1.05 pu
- Normal operation: 0.95-1.05 pu

Provide your prediction for ALL 118 buses."""

    result = call_llm(system_prompt, user_prompt, temperature=0.1, max_tokens=2000)
    
    with lock:
        print(f"  Completed: {idx}")
    
    return idx, result

# Run predictions in parallel
print(f"Running {len(test_features)} test samples in parallel...")
print("(Using ThreadPoolExecutor for concurrent API calls)\n")

predictions = {}
lock = threading.Lock()

# Prepare arguments for each sample
test_list = [(idx, row, analysis, lock) for idx, row in test_features.iterrows()]

# Use ThreadPoolExecutor for parallel API calls
# Adjust max_workers based on API rate limits
with concurrent.futures.ThreadPoolExecutor(max_workers=10) as executor:
    results = executor.map(predict_single_sample, test_list)
    
    for idx, result in results:
        if result:
            predictions[str(idx)] = result

print(f"\nCompleted {len(predictions)} predictions!")

# Save raw predictions
with open(RESULTS_DIR / "phase2_predictions.json", "w") as f:
    json.dump(predictions, f, indent=2)

print(f"Saved to {RESULTS_DIR / 'phase2_predictions.json'}")
```

```python
# Parse ALL predictions and calculate MAE/RMSE
print("=== Parsing ALL LLM Predictions ===\n")

parsed_predictions = {}
parse_errors = 0

for idx, pred_text in predictions.items():
    parsed = parse_grid_predictions(pred_text)
    if parsed:
        parsed_predictions[idx] = parsed
    else:
        parse_errors += 1

print(f"Successfully parsed: {len(parsed_predictions)}/{len(predictions)} predictions")
print(f"Parse errors: {parse_errors}")

# Calculate MAE and RMSE for ALL predictions
if parsed_predictions:
    all_mae = []
    all_rmse = []
    bus_errors = {}
    
    for idx, pred in parsed_predictions.items():
        # Use timestamp key directly - idx is the timestamp string
        if idx not in test_targets.index:
            continue
        
        actual = test_targets.loc[idx].to_dict()
        
        errors = []
        for bus in actual.keys():
            if bus in pred:
                err = abs(actual[bus] - pred[bus])
                errors.append(err ** 2)
                if bus not in bus_errors:
                    bus_errors[bus] = []
                bus_errors[bus].append(err)
        
        if errors:
            mae = np.mean([abs(actual[bus] - pred[bus]) for bus in actual.keys() if bus in pred])
            rmse = np.sqrt(np.mean(errors))
            all_mae.append(mae)
            all_rmse.append(rmse)
    
    if all_mae:
        overall_mae = np.mean(all_mae)
        overall_rmse = np.sqrt(np.mean([e**2 for e in all_mae]))
        
        # Calculate MAPE (Mean Absolute Percentage Error)
        all_mape = []
        for idx, pred in parsed_predictions.items():
            if idx not in test_targets.index:
                continue
            actual = test_targets.loc[idx].to_dict()
            # MAPE = (1/n) * sum(|actual - pred| / |actual|) * 100
            pct_errors = [abs(actual[bus] - pred[bus]) / abs(actual[bus]) * 100 
                         for bus in actual.keys() if bus in pred and actual[bus] != 0]
            if pct_errors:
                all_mape.append(np.mean(pct_errors))
        
        mape = np.mean(all_mape) if all_mape else None
        
        print(f"\n=== LLM Prediction Performance (ALL {len(all_mae)} samples) ===")
        print(f"  MAE:  {overall_mae:.4f} pu")
        print(f"  RMSE: {overall_rmse:.4f} pu")
        if mape:
            print(f"  MAPE: {mape:.2f}%")
        
        # Per-bus analysis
        if bus_errors:
            bus_mae = {bus: np.mean(errs) for bus, errs in bus_errors.items()}
            worst_bus = max(bus_mae, key=bus_mae.get)
            best_bus = min(bus_mae, key=bus_mae.get)
            print(f"\n  Best bus:  {best_bus} ({bus_mae[best_bus]:.4f} pu)")
            print(f"  Worst bus: {worst_bus} ({bus_mae[worst_bus]:.4f} pu)")
        
        # Compare with XGBoost
        print(f"\n=== Comparison with XGBoost ===")
        try:
            with open(RESULTS_DIR / "xgboost_grid_results.json") as f:
                xgb_results = json.load(f)
            xgb_mae = xgb_results.get("mae_overall", 0)
            xgb_rmse = xgb_results.get("rmse_overall", 0)
            print(f"  XGBoost MAE: {xgb_mae:.4f} pu")
            print(f"  LLM MAE:     {overall_mae:.4f} pu")
            print(f"  Ratio: LLM is {overall_mae/xgb_mae:.1f}x worse than XGBoost")
            if mape:
                print(f"  MAPE: {mape:.2f}%")
            print(f"\n  XGBoost RMSE: {xgb_rmse:.4f} pu")
            print(f"  LLM RMSE:     {overall_rmse:.4f} pu")
        except Exception as e:
            print(f"  Could not load XGBoost results: {e}")
else:
    print("No valid predictions to evaluate")
    overall_mae = None
    overall_rmse = None
```

```python
# Phase 4: Calculate Replicability Score with ALL samples
print("=== Phase 4: Replicability Test ===\n")

if not parsed_predictions or not rulebook:
    print("Missing predictions or rulebook. Skipping replicability test.")
    replicability_results = None
else:
    replicability_results = {
        "domain": "grid",
        "samples": [],
        "original_llm_performance": {},
    }
    
    # Calculate original LLM MAE for ALL samples
    original_maes = []
    for idx, pred in parsed_predictions.items():
        if idx not in test_targets.index:
            continue
        
        actual = test_targets.loc[idx].to_dict()
        
        errors = [abs(actual[bus] - pred[bus]) for bus in actual.keys() if bus in pred]
        
        if errors:
            original_maes.append(np.mean(errors))
    
    if original_maes:
        overall_mae = np.mean(original_maes)
        replicability_results["original_llm_performance"] = {
            "mae_overall": float(overall_mae),
            "rmse_overall": float(np.sqrt(np.mean([e**2 for e in original_maes]))),
            "n_samples": len(original_maes)
        }
        print(f"Original LLM MAE: {overall_mae:.4f} pu ({len(original_maes)} samples)")
    
    # Test replicability on ALL samples (in parallel)
    print(f"\nTesting replicability on {len(parsed_predictions)} samples...")
    
    test_sample_indices = list(parsed_predictions.keys())
    replicable_count = 0
    fresh_maes = []
    
    def test_replicability(idx):
        if idx not in test_features.index or idx not in test_targets.index:
            return None, False
        
        test_row = test_features.loc[idx]
        actual = test_targets.loc[idx].to_dict()
        
        # Format features
        features_str = "\n".join([f"{col}: {test_row[col]:.4f}" for col in test_features.columns[:15]])
        
        system_prompt = f"""You are a power systems engineer. Use this Decision Rulebook to predict voltages:

{rulebook[:2500]}"""
        
        user_prompt = f"""Given these system features:
{features_str}

Predict voltage magnitudes (vm_1 to vm_118) in pu. Format as:
Bus 1: X.XXX
Bus 2: X.XXX
..."""
        
        fresh_response = call_llm(system_prompt, user_prompt, temperature=0.1, max_tokens=2000)
        
        if fresh_response:
            fresh_parsed = parse_grid_predictions(fresh_response)
            if fresh_parsed:
                errors = [abs(actual[bus] - fresh_parsed[bus]) for bus in actual.keys() if bus in fresh_parsed]
                if errors:
                    return np.mean(errors), True
        
        return None, False
    
    # Run in parallel
    with concurrent.futures.ThreadPoolExecutor(max_workers=5) as executor:
        results = list(executor.map(test_replicability, test_sample_indices))
    
    for idx, (fresh_mae, success) in zip(test_sample_indices, results):
        if success and fresh_mae:
            fresh_maes.append(fresh_mae)
            orig_pred = parsed_predictions.get(idx, {})
            if orig_pred and idx in test_targets.index:
                actual = test_targets.loc[idx].to_dict()
                orig_mae = np.mean([abs(actual[bus] - orig_pred[bus]) for bus in actual.keys() if bus in orig_pred])
                if fresh_mae <= orig_mae * 1.2:  # Within 20%
                    replicable_count += 1
    
    # Calculate replicability score
    if fresh_maes:
        replicability_score = replicable_count / len(fresh_maes)
        replicability_results["replicability_score"] = float(replicability_score)
        replicability_results["samples_tested"] = len(fresh_maes)
        replicability_results["samples_replicable"] = replicable_count
        replicability_results["fresh_llm_mae"] = float(np.mean(fresh_maes))
        
        print(f"\n=== Replicability Results ===")
        print(f"  Score: {replicability_score:.2%}")
        print(f"  {replicable_count}/{len(fresh_maes)} samples replicable")
        print(f"  Fresh LLM MAE: {np.mean(fresh_maes):.4f} pu")
    else:
        print("Could not calculate replicability")
        replicability_results = None
```

## 6. Phase 3: Decision Rulebook

```python
def run_phase3(call_llm, predictions, test_targets, analysis, domain='grid'):
    """Phase 3: Extract Decision Rulebook."""
    print("\n" + "="*60)
    print("PHASE 3: RULEBOOK EXTRACTION")
    print("="*60)

    if not predictions:
        print("No predictions available.")
        return None

    comparison = ""
    
    # Get list of prediction indices
    pred_indices = list(predictions.keys())#[:5]  # Use first 5 for rulebook
    
    for idx in pred_indices:
        comparison += f"\n## Test Case (index={idx})\n"
        comparison += f"Predicted (from Phase 2):\n{predictions[idx][:400]}\n\n"
        
        # Get actual values - handle both string and int keys
        try:
            if idx in test_targets.index:
                actual = test_targets.loc[idx].to_dict()
            else:
                # Try by position if string key doesn't exist
                idx_pos = list(test_targets.index).index(idx) if idx in list(test_targets.index) else 0
                actual = test_targets.iloc[idx_pos].to_dict()
        except:
            actual = {}
        
        if actual:
            comparison += f"Actual:\n"
            for bus, val in list(actual.items())[:5]:
                comparison += f"  {bus}: {val:.4f}\n"
            comparison += "\n"

    # Build the rulebook extraction prompt
    system_prompt = """You are an expert power systems engineer. Your task is to extract DECISION RULES from your predictions."""

    user_prompt = f"""## Your Previous Predictions

{comparison}

## Your Analysis (from Phase 1)

{analysis[:2000]}

## Your Task

Based on your predictions above and your analysis, create a DECISION RULEBOOK that explains how you predict voltages. Format as:

### Rule Category 1
- IF [condition] THEN [voltage prediction]
- IF [condition] THEN [voltage prediction]

### Rule Category 2
...

Focus on:
1. Load conditions → voltage relationships
2. Generation mix → voltage relationships  
3. Time-of-day patterns
4. Bus categories (generator vs load)

Extract 5-10 clear rules that could be used by another engineer to predict voltages without running a full power flow simulation."""

    print("Extracting decision rules from predictions...")
    rulebook = call_llm(system_prompt, user_prompt, temperature=0.3, max_tokens=20000)

    if rulebook:
        print(f"Rulebook extracted! Length: {len(rulebook)} chars")
        with open(RESULTS_DIR / "phase3_rulebook.txt", "w") as f:
            f.write(rulebook)
        print(f"Saved to: {RESULTS_DIR / 'phase3_rulebook.txt'}")
    else:
        print("WARNING: Rulebook extraction returned empty!")
        rulebook = "Could not extract rulebook"

    return rulebook
```

```python
# Run Phase 3
rulebook = run_phase3(call_llm, predictions, test_targets, analysis)
```

```python
print("=== Phase 3: Decision Rulebook ===\n")
print(rulebook if rulebook else "No rulebook available")
print("\n...")
```

## 7. Phase 4: Replicability Test

```python
def calculate_replicability(call_llm, predictions, test_targets, rulebook, test_features, n_samples=5):
    """Phase 4: Calculate Replicability Score.

    The Replicability Score answers: Can a FRESH LLM, given ONLY the Rulebook
    and test features, reproduce predictions that match the ACTUAL ground truth?
    """
    print("\n" + "="*60)
    print("PHASE 4: REPLICABILITY TEST")
    print("="*60)

    if not rulebook or not predictions:
        print("Missing rulebook or predictions. Skipping.")
        return None

    results = {"samples": [], "domain": "grid"}

    all_original = []
    all_fresh = []
    all_actual = []

    test_indices = list(predictions.keys())[:n_samples]

    feature_descriptions = {
        'hour': 'Hour of day (0-23)',
        'load_p_r1_mw': 'R1 load (MW)',
        'gen_solar_p_r1_mw': 'R1 solar (MW)',
        'gen_wind_p_r1_mw': 'R1 wind (MW)',
        'gen_hydro_p_r1_mw': 'R1 hydro (MW)',
        'gen_thermal_p_r1_mw': 'R1 thermal (MW)',
        'slack_vm_pu': 'Slack bus voltage (pu)',
    }

    for idx in test_indices:
        idx_int = int(idx)
        print(f"\nSample {idx}...")

        original_response = predictions.get(idx, '')
        original_voltages = parse_grid_predictions(original_response)
        test_row = test_features.iloc[idx_int] if idx_int < len(test_features) else None
        actual_row = test_targets.iloc[idx_int].to_dict() if idx_int < len(test_targets) else {}

        if test_row is None:
            print(f"  No test row for index {idx_int}. Skipping.")
            continue

        # Format features
        features_str = "\n".join(
            f"  {col} ({feature_descriptions.get(col, col)}): {test_row[col]:.4f}"
            for col in train_features.columns if col in test_row.index
        )

        # Fresh LLM call
        replication_system = (
            "You are a power systems engineer. Use the Decision Rulebook below "
            "to predict voltage magnitudes at each bus for the given test case. "
            "Output ONLY the predictions in the format:\n"
            "Bus: <number>\n"
            "Predicted vm_pu: <value>"
        )

        replication_user = f"""## Decision Rulebook

{rulebook}

## Test Case Features

{features_str}

## Your Task

Using ONLY the rules above and the test features, predict the voltage magnitude "\
"(vm_pu) at each bus. Output in this format:
Bus: <number>
Predicted vm_pu: <value>
..."
"""

        fresh_response = call_llm(replication_system, replication_user, temperature=0.1, max_tokens=1024)

        if not fresh_response:
            print(f"  Fresh LLM call failed. Skipping.")
            continue

        fresh_voltages = parse_grid_predictions(fresh_response)

        # Calculate MAE
        if isinstance(original_voltages, dict) and isinstance(fresh_voltages, dict):
            matching_buses = set(original_voltages.keys()) & set(fresh_voltages.keys())
            if matching_buses:
                errors_original = []
                errors_fresh = []
                for bus in matching_buses:
                    orig = original_voltages[bus]
                    fresh = fresh_voltages[bus]
                    actual = actual_row.get(bus, None)
                    if actual is not None:
                        errors_original.append(abs(orig - actual))
                        errors_fresh.append(abs(fresh - actual))

                if errors_original and errors_fresh:
                    mae_original = float(np.mean(errors_original))
                    mae_fresh = float(np.mean(errors_fresh))
                    replicability_score = max(0.0, 1.0 - mae_fresh / mae_original) if mae_original > 0 else 0.0

                    print(f"  Original MAE: {mae_original:.4f}, Fresh MAE: {mae_fresh:.4f}")
                    print(f"  Replicability Score: {replicability_score:.2%}")

                    results["samples"].append({
                        "index": idx,
                        "n_matching_buses": len(matching_buses),
                        "mae_original": mae_original,
                        "mae_fresh": mae_fresh,
                        "replicability_score": replicability_score,
                    })

                    for bus in matching_buses:
                        all_original.append(original_voltages[bus])
                        all_fresh.append(fresh_voltages[bus])
                        actual = actual_row.get(bus)
                        if actual is not None:
                            all_actual.append(actual)

        time.sleep(0.5)

    # Overall results
    if all_original and all_fresh and all_actual:
        mae_original_all = float(np.mean([abs(o - a) for o, a in zip(all_original, all_actual)]))
        mae_fresh_all = float(np.mean([abs(f - a) for f, a in zip(all_fresh, all_actual)]))
        results["overall_mae_original"] = mae_original_all
        results["overall_mae_fresh"] = mae_fresh_all
        results["replicability_score"] = max(0.0, 1.0 - mae_fresh_all / mae_original_all) if mae_original_all > 0 else 0.0
        results["n_total_buses"] = len(all_original)

        print(f"\n=== Grid Replicability Results ===")
        print(f"  Original LLM MAE: {mae_original_all:.4f} pu")
        print(f"  Fresh LLM MAE:    {mae_fresh_all:.4f} pu")
        print(f"  Replicability Score: {results['replicability_score']:.2%}")

    return results
```
